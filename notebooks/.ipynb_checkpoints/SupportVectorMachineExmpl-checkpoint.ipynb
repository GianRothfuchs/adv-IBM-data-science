```python
# for schema generation
from pyspark.sql.types import StructType,StructField, IntegerType
# for adding string to df
from pyspark.sql.functions import lit


import os
file_list = os.listdir('HMP_Dataset')
file_list_filtered = [s for s in file_list if '_' in s]
```

```python
#init DataFrame

df = None
# this schema expects three columns with integers
schema = StructType([
    StructField("x",IntegerType(),True),
    StructField("y",IntegerType(),True),
    StructField("z",IntegerType(),True)])

for category in file_list_filtered:
    data_files = os.listdir('HMP_Dataset/' + category)
    
    for data_file in data_files:
        print(data_file)
        #creating a temporary dataframe from file
        temp_df = spark.read.option("header","false").option("delimiter"," ").csv('HMP_Dataset/' + category + '/' + data_file, schema=schema)
        # adding source file and contextual info about the data file: df.withColumn() used to create rename or change column
        temp_df = temp_df.withColumn('class',lit(category))
        temp_df = temp_df.withColumn('source',lit(data_file))
        
        if df is None:
            df = temp_df
        else:
            # df.union appends data vertically
            df = df.union(temp_df)
            
        
        
```

```python
df.show()
```

```python
df.createOrReplaceTempView('df')
```

```python
# limitiing to two factors as model will be binary classifier
df_two_classes = spark.sql("select * from df where class in ('Use_telephone','Standup_chair')")
splits = df_two_classes.randomSplit([0.8,0.2])

#splits = df.randomSplit([0.8,0.2])

df_train = splits[0]
df_test = splits[1]
```

```python

from pyspark.ml import Pipeline

from pyspark.ml.feature import StringIndexer
from pyspark.ml.feature import OneHotEncoder
from pyspark.ml.linalg import Vectors
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.feature import Normalizer

from pyspark.ml.classification import LinearSVC

# Pipeline: data pre-processing
indexer = StringIndexer(inputCol="class", outputCol="label")
encoder = OneHotEncoder(inputCol='label', outputCol='labelVec')
vectorAssembler = VectorAssembler(inputCols=["x","y","z"], outputCol = "features")
normalizer = Normalizer(inputCol='features',outputCol='features_norm',p=1.0)

#Pipeline: model Support Vector Machine
lsvc = LinearSVC(maxIter=10,regParam=0.1)

# Pipeline: line up
pipeline = Pipeline(stages=[indexer,encoder,vectorAssembler,normalizer,lsvc])
```

```python
# model fit on training data (takes a while)
model = pipeline.fit(df_train)
```

```python
# model evaluation
from pyspark.ml.evaluation import BinaryClassificationEvaluator

prediction = model.transform(df_train)

evaluator = BinaryClassificationEvaluator(rawPredictionCol='rawPrediction') # rawPReidction is score variable created by LinearSVC

# accuracy for training set
evaluator.evaluate(prediction)

# accuracy for test set
prediction = model.transform(df_test)
evaluator.evaluate(prediction)


```
